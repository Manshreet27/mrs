```python

import ast



# Define the convert function
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L

# Apply the convert function to the 'genres' column
movies['genres'] = movies['genres'].apply(convert)

# Display the updated DataFrame
(movies.head())
```

```python
# to get the orignal name of charater in the movie
def convert3(obj):
    L = []
    counter=0
    for i in ast.literal_eval(obj):
        if counter !=3:
            L.append(i['name'])
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
# to finc only those crew whose job role is director
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
# we have total 4806 movies
#and 4806 vector
# each vector have 5000 numbers
# each movie is vector so each movie is used to calculate distance with all other movie
# max distance similarity less
# we will not calculate euclidiean distance we will calculate cosine distance
# because euclidiean distance is tip to tip distance so we dont have to calulate that we need to calculte the angle between them 
# as eg: bteween two movie if angle is :0 so same , 5 so less distance ,10 some more distance, 90 there is large distance so more disimilarity and if 180 so full on opposite movie
# if we work on high dimmensional data(5000) eucliedian distace most of the time fails
#so thats why we use cosine ditance -- threta between two vector
# distance inversle propotional to similarity 
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6] #as 1st movie similarity with own shoul be 1 so thats why it is showing 1 in first place same goes with 2nd movie so on the second place will be 1
# sorted and in the form of list
# it gives you the most related 5 movies 
```

```python
# if someone give the movie it should return the similar 5 movie 
def recommend(movie):
    movie_index=newdf[newdf['title']== movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
# it will give the 5 related movies 
    for i in movies_list:
        print(newdf.iloc[i[0]].title) # it will print the title of the movies related to the selected movie
         # it will print all the movies list 

# if we print i only so we sill get id 
```

```python
recommend('Batman Begins')
```
